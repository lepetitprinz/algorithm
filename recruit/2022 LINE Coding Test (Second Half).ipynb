```python
from collections import deque


def bfs(graph, start, goal, visited, width, height, move_list):
    r, c, cnt = start
    queue = deque([[r, c, cnt]])
    visited[r][c] = 0

    while queue:
        r, c, cnt = queue.popleft()

        for kind, moves in move_list.items():
            for move in moves:
                if check_move(r, c, kind, move, graph, width, height):
                    nr = r + move[0]
                    nc = c + move[1]
                    if nr == goal[0] and nc == goal[1]:
                        return cnt + 1
                    
                    if not visited[nr][nc]:
                        queue.append([nr, nc, cnt + 1])
                        visited[nr][nc] = 1

    return -1


def check_move(r, c, kind, move, graph, width, height):
    flag = True
    if (0 <= r + move[0] < height) and (0 <= c + move[1] < width):
        if kind == 1:
            flag = check_move1(r, c, move, graph)
        elif kind == 2:
            flag = check_move2(r, c, move, graph)
        elif kind == 3:
            flag = check_move3(r, c, move, graph)
        elif kind == 4:
            flag = check_move4(r, c, move, graph)
    else:
        flag = False

    return flag


def check_move1(r, c, move, graph):
    flag = False
    if graph[r + move[0]][c + move[1]] == 'H':
        flag = True

    return flag


def check_move2(r, c, move, graph):
    flag = True
    if graph[r][c + move[1]] == 'H':
        for i in range(c + 1, c + move[1]):
            if graph[r][i] != '.':
                flag = False
                break
        for j in range(c, c + move[1] + 1):
            if graph[r - 1][j] != '.':
                flag = False
                break
    else:
        flag = False

    return flag


def check_move3(r, c, move, graph):
    flag = True
    if graph[r + move[0]][c] == 'H':
        if graph[r + move[0] + 1][c] != '.':
            flag = False
    else:
        flag = False

    return flag


def check_move4(r, c, move, graph):
    flag = True
    if graph[r + move[0]][c + move[1]] == 'H':
        if move[1] == 1:
            if graph[r][c + 1] != '.':
                flag = False
            if graph[r - 1][c] != '.':
                flag = False
        else:
            if graph[r][c - 1] != '.':
                flag = False
            if graph[r - 1][c] != '.':
                flag = False
    else:
        flag = False

    return flag


def solution(wall):
    graph = [list(w) for w in wall]
    height = len(graph)
    width = len(graph[0])

    goals = []
    for r in range(height):
        for c in range(width):
            if graph[r][c] == 'H':
                goals.append((r, c))
                

    move_list = {
        1: [(1, 0), (0, 1), (-1, 0), (0, -1)],
        2: [(0, 2), (0, -2), (0, 3), (0, -3)],
        3: [(-2, 0)],
        4: [(-1, 1), (-1, -1)]
    }
    
    result = [[0] * width for _ in range(height)]
    for r, c in goals:
        visited = [[False] * width for _ in range(height)]
        cnt = bfs(graph, [height - 1, 0, 1], [r, c], visited, width, height, move_list)
        result[r][c] = cnt
        
    result[height-1][0] = 1
    
    return result
```
